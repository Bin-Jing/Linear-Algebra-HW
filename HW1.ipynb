```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

# 三個線性系統

```python
x1 = np.linspace(0,5,2)
y1 = 3*x1 + 1
plt.plot(x1,y1)
x2 = np.linspace(0,5,2)
y2 = 3*x2 - 1
plt.plot(x2,y2)
plt.title("No solution")
plt.show()
```

```python
x3 = np.linspace(0,5,2)
y3 = 10*x3
plt.plot(x3,y3)
x4 = np.linspace(0,5,2)
y4 = x4 + 10
plt.plot(x4,y4)
plt.title("One solution")
plt.show()
```

```python
x5 = np.linspace(0,5,2)
y5 = 87*x5 + 65536
plt.plot(x5,y5)
x6 = np.linspace(0,5,2)
y6 = 87*x5 + 65536
plt.plot(x6,y6)
plt.title("Infinitely solution")
plt.show()
```
