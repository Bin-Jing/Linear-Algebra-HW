```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
def getpoint(x):
    return np.sin(9*x)*np.cos(7*x)
def drawfig(x):
    plt.plot(x,np.sin(9*x)*np.cos(7*x),'ro')
    a = np.linspace(0, 6.5, 1200)
    plt.plot(a,np.sin(9*a)*np.cos(7*a))
```

```python
rate = 0.001
arr = np.array([])
```

# 從頭掃到尾方法
### 適用於範圍較小的

```python
minx = 0
x = np.linspace(0, 6.5, 1200)
y = np.sin(9*x)*np.cos(7*x)
plt.xlim(0,6.5)
plt.plot(x,y)
for i in range(0,(int)(6.5/rate)):
    arr = np.append(arr,getpoint(i*rate)) 
    if(np.min(arr) >= getpoint(i*rate)):
        minx = i
drawfig(minx*rate)
print("從頭找到尾得最小值 = {}".format(arr[minx]))
```

# 梯度找法

```python
import random
        
```

```python
def func(x):
    return np.sin(9*x)*np.cos(7*x)
def error(x):
    return (func(x)-(-1))**2#最小值為-1
def derivative_descent(x):
    delta = rate
    derivative = (error(x + delta) - error(x - delta)) / (delta * 2)

    alpha = 0.001
    x = x - derivative * alpha
    return x
```

```python
arr2 = np.array([])
arr3 = np.array([])
minx = 0
for i in range(0,100):
    a = random.randint(0,650)/100
    arr2 = np.append(arr2,a)
    for i in range(100):
        x = derivative_descent(a)
        arr3 = np.append(arr3,func(x))
        if(np.min(arr3) >= func(x)):
            minx = x
            
print("梯度找到的最小值 = {}".format(np.min(arr3)))
drawfig(minx)
```
