```python
c = np.linspace(0,2*np.pi,100)
r = 1+15*np.sin(c)
x = r*np.cos(c)
y = r*np.sin(c)

c1 = np.linspace(0,2*np.pi,100)
r1 = 1-np.sin(c1)
x1 = r1*np.cos(c1)
y1 = r1*np.sin(c1)+3

c2 = np.linspace(0,2*np.pi,100)
r2 = 1
x2 = r2*np.cos(c2)-3
y2 = r2*np.sin(c2)+10

c3 = np.linspace(0,2*np.pi,100)
r3 = 1
x3 = r3*np.cos(c2)+3
y3 = r3*np.sin(c2)+10

x4 = np.linspace(0,0.2,100)
y4 = (-12)*x4+8
```

```python
plt.plot(x,y)
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(x4,y4)
```

```python
A = np.array([[3,13],[1,-23]])
xy  = np.array([x,y])
Axy = A@xy
xy1  = np.array([x1,y1])
Axy1 = A@xy1
xy2  = np.array([x2,y2])
Axy2 = A@xy2
xy3  = np.array([x3,y3])
Axy3 = A@xy3
xy4  = np.array([x4,y4])
Axy4 = A@xy4
plt.plot(Axy[0], Axy[1])
plt.plot(Axy1[0], Axy1[1])
plt.plot(Axy2[0], Axy2[1])
plt.plot(Axy3[0], Axy3[1])
plt.plot(Axy4[0], Axy4[1])
```
